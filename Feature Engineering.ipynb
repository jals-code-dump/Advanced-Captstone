# Feature Engineering

This notebook does the following:

* Takes in noisy ETL documents
* Uses Binary thresholding and OTSU to convert to binary

### Dependencies

```python
import cv2
import os
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (20, 10)
```

### Thresholding

* Takes noisy documents
* Outputs originals and new versions

```python
filepath = r"C:\GitHub\mf724-ocr\ETL Documents\\" 
docs = []

for i in os.listdir(filepath):
    image = cv2.imread(filepath + i)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(image,0,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)
    docs.append([image, thresh])
```

### Results

```python
plt.figure()
plt.imshow(docs[0][0], cmap="gray")
plt.figure()
plt.imshow(docs[0][1], cmap="gray")
```

### Output

```python
for i, v in enumerate(docs):
    filename = r"C:\GitHub\mf724-ocr\Denoised\\" + str(i) + ".jpg"
    cv2.imwrite(filename, v[1])
```
