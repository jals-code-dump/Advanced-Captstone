```python
import os
import numpy as np
from scipy import signal
from PIL import Image
from matplotlib import pyplot as plt
import cv2
```

```python
input_path = "C:\\GitHub\\mf724-ocr\\ETL Documents\\"
output_path = "C:\\GitHub\mf724-ocr\\Denoised\\"

# store image names in list for later use
scanned_img = sorted(os.listdir(input_path))
print(scanned_img)
```

```python
for i, img in enumerate(scanned_img):
    image = cv2.imread(input_path + img)
    image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    image = np.asarray(image)/255.0
    # estimate 'background' color by a median filter
    bg = signal.medfilt2d(image, 11)
    # compute 'foreground' mask as anything that is significantly darker than
    # the background
    mask = image < bg - 0.1
    # return the input value for all pixels in the mask or pure white otherwise
    mask = np.where(mask, image, 1.0)
    tmp = np.asarray(mask*255.0, dtype=np.uint8)
    plt.imshow(tmp)
    op = Image.fromarray(tmp)
    plt.imshow(op)
    cv2.imwrite()
```
